```python
new_dic = { word:idx+3 for word,idx in dictionary.items()}
```

```python
# just like training data, we should also process the testing data

xtest_padded = keras.utils.pad_sequences(xtest, 
                                         maxlen=HP_MAX_LEN, 
                                         padding='post',
                                         truncating='post',
                                         )
```

```python
# PATTERN DETECTION -> at least one input layer, output layer, and at least 1  HIDDEN LAYER

# these layers, also known as FULLY CONNECTED LAYERS or DENSE LAYERS accept only sequential numbers

# NLP PROBLEM -> EMBEDDING -> our words as numbers will be converted into multi-dimensional vectors

# sentence length of 100, and a total of 1000 possible english words
# for every word, at every position, i will generate a vector matrix 

# i am the batman
# the batman is hungry
# hunger is batman 

# with 4 positions and 4 words   -> This matrix is Created by EMBEDDING LAYER
#.      pos1 pos2 pos3 pos4
# word1  v1.  v2.  v3   v4
# word2  v5.  v6.  v7.  v8
# word3  v9.  v10. v11. v12
# word4  v13  v14  v15  v16

# But dense layers do not accept matrix, they accept only linear numbers!

# we need a way in which this matrix should convert into a single line of vectors!

# just calculate average of all the vectors!

# OVERALL NETWORK NOW LOOKS LIKE->

# Embedding Layer -> GlobalAveragePooling1D -> Pattern_Recognition
# Embedding Layer -> GlobalAveragePooling1D -> (Dense Layers) X 3

```

```python
l1 = keras.layers.Embedding(10000, 32)  # number of parameters = 10k X 32
l2 = keras.layers.GlobalAveragePooling1D() # number of parameters = 0
l3 = keras.layers.Dense(64, activation='relu') # in order to move away from straight lines, 
l4 = keras.layers.Dense(128, activation='relu') # we need to use 'ACTIVATION FUNCTIONS'
l5 = keras.layers.Dense(1, activation = 'sigmoid')

# relu -> removes all negative numbers from input
# sigmoid -> single probability, between 0 to 1
# softmax -> converts results into proabability distribution 
```

```python
model = keras.Sequential([l1,l2, l3, l4, l5])
model.compile(loss='binary_crossentropy' , optimizer='adam')
```

```python
model.fit(xtrain_padded, ytrain, epochs=20)
```

```python
# problem-> too many words which neural network cannot figure out within its 10,000 limit!
# solution -> read your data only considering the top 10,000 words!
# rest of the words will become UNKNOWN for us, which is handled by using 2
data = imdb.load_data(num_words=10000)
```

```python
# let's do the entire processing again only considering the top 10k words

(xtrain,ytrain), (xtest, ytest) = data
xtrain_padded = keras.utils.pad_sequences(xtrain, 
                                         maxlen=HP_MAX_LEN, 
                                         padding='post',
                                         truncating='post',
                                         )
xtest_padded = keras.utils.pad_sequences(xtest, 
                                         maxlen=HP_MAX_LEN, 
                                         padding='post',
                                         truncating='post',
                                         )
```

```python
model.fit(xtrain_padded, ytrain, epochs=20)
```

```python
p = model.predict(xtest_padded)
```

```python
classifier = lambda val: 1 if val>0.5 else 0
prob = [ classifier(result) for result in p]
```

```python
correct = 0
wrong = 0

for i in range(len(ytest)):
    if(prob[i] == ytest[i]):
        correct = correct+1
    else:
        wrong = wrong + 1
```
